```python
import glob, os, sys
import math

import numpy as np
import pandas as pd
import xarray as xr
import scipy.interpolate

import matplotlib.pyplot as plt

from icenet2.model.models import linear_trend_forecast
```

```python
x = np.linspace(0, 6*np.pi, 20)
y = np.sin(x) * 2
```

```python
xi = np.linspace(0, 6*np.pi, 1000) - 10
inter = scipy.interpolate.interp1d(x, y, kind='quadratic')
```

```python
yi = inter(xi)
```

```python
plt.plot(x, y, 'gx')
plt.plot(xi, yi, '-')
plt.show()
```

```python
X, Y = np.meshgrid(xi, yi)
def z_func(x, y): 
    return (np.cos(x) * y) - (np.cos(y) * x)
Z = np.array([z_func(x, y) for x, y in zip(np.ravel(X), np.ravel(Y))])
```

```python
print(X.shape)
print(Y.shape)
Z = Z.reshape(X.shape)
print(Z.shape)
```

```python
plt.pcolormesh(X, Y, Z)
plt.show()
```
